```python
import numpy as np
from numpy import nan as NA
import pandas as pd

sample_data_frame = pd.DataFrame(np.random.rand(10, 4))

sample_data_frame.iloc[1,0] = NA
sample_data_frame.iloc[2,2] = NA
sample_data_frame.iloc[5:,3] = NA

sample_data_frame
```

```python
sample_data_frame.dropna() # 리스트 와이즈
```

```python
sample_data_frame[[0,1,2]].dropna() # 페어 와이즈
```

```python
# 결측치 보완
import numpy as np
from numpy import nan as NA
import pandas as pd

sample_data_frame = pd.DataFrame(np.random.rand(10, 4))

sample_data_frame.iloc[1,0] = NA
sample_data_frame.iloc[2,2] = NA
sample_data_frame.iloc[5:,3] = NA

sample_data_frame
```

```python
sample_data_frame.fillna(0) # 0으로 채우기
```

```python
# 앞에있는 데이터로 채우기
sample_data_frame.ffill()
```

```python
# 평균값 대입
import numpy as np
from numpy import nan as NA
import pandas as pd

sample_data_frame = pd.DataFrame(np.random.rand(10, 4))

sample_data_frame.iloc[1,0] = NA
sample_data_frame.iloc[2,2] = NA
sample_data_frame.iloc[5:,3] = NA

# fillna로 nan 부분에 열의 평균값을 대입
sample_data_frame.fillna(sample_data_frame.mean())
```
